```python
FILE_PATH = "python_notes.txt"

# ----------------------
# Part A: Write to File
# ----------------------

def write_initial_notes():
    with open(FILE_PATH, "w", encoding="utf-8") as file:
        file.write("Topic 1: Variables store data. Python is dynamically typed.\n")
        file.write("Topic 2: Lists are ordered and mutable.\n")
        file.write("Topic 3: Dictionaries store key-value pairs.\n")
        file.write("Topic 4: Loops automate repetitive tasks.\n")
        file.write("Topic 5: Exception handling prevents crashes.\n")
    print("File written successfully.")

def append_extra_notes():
    with open(FILE_PATH, "a", encoding="utf-8") as file:
        file.write("Topic 6: Functions help reuse code.\n")
        file.write("Topic 7: APIs connect applications.\n")
    print("Lines appended.")

write_initial_notes()
append_extra_notes()

# ----------------------
# Part B: Read from File
# ----------------------

with open(FILE_PATH, "r", encoding="utf-8") as file:
    lines = file.readlines()

print("\nFile Contents:")
for number, text in enumerate(lines, start=1):
    print(f"{number}. {text.strip()}")

total_lines = len(lines)
print(f"\nTotal lines: {total_lines}")

search_term = input("\nEnter a keyword to search in the notes: ").lower()

print("\nSearch Results:")
match_found = False
for text in lines:
    if search_term in text.lower():
        print(text.strip())
        match_found = True

if not match_found:
    print("No matching lines found for that keyword.")
```

```python
import requests

BASE_URL = "https://dummyjson.com/products"


def fetch_products(limit=20):
    url = f"{BASE_URL}?limit={limit}"
    r = requests.get(url)
    r.raise_for_status()
    return r.json()["products"]


def print_products_table(products):
    print(f"{'ID':<4}| {'Title':<32}| {'Category':<15}| {'Price':<9}| Rating")
    print("-" * 80)
    for p in products:
        print(
            f"{p['id']:<4}| "
            f"{p['title']:<32}| "
            f"{p['category']:<15}| "
            f"${p['price']:<8.2f}| {p['rating']}"
        )


def filter_and_sort_by_rating(products, min_rating=4.5):
    result = [p for p in products if p["rating"] >= min_rating]
    return sorted(result, key=lambda x: x["price"], reverse=True)


def fetch_by_category(category):
    url = f"{BASE_URL}/category/{category}"
    r = requests.get(url)
    r.raise_for_status()
    return r.json()["products"]


def create_product(payload):
    url = f"{BASE_URL}/add"
    r = requests.post(url, json=payload)
    r.raise_for_status()
    return r.json()


# ---------------------- Step 1 ----------------------
print("Fetching 20 products...\n")
all_products = fetch_products(20)
print_products_table(all_products)

# ---------------------- Step 2 ----------------------
print("\nFiltered Products (Rating >= 4.5):\n")
high_rated = filter_and_sort_by_rating(all_products, 4.5)
for p in high_rated:
    print(f"{p['title']} | Price: ${p['price']:.2f} | Rating: {p['rating']}")

# ---------------------- Step 3 ----------------------
print("\nLaptops Category:\n")
laptops = fetch_by_category("laptops")
for lp in laptops:
    print(f"{lp['title']} | Price: ${lp['price']:.2f}")

# ---------------------- Step 4 ----------------------
print("\nSending POST Request...\n")
payload = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API",
}
response_data = create_product(payload)

print("Response from server:")
print(response_data)
```

```python
import requests

BASE_URL = "https://dummyjson.com/products"

# ----------------------
# Part A: Safe Divide
# ----------------------
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"


print("Safe Divide Tests:")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))


# ----------------------
# Part B: Safe File Reader
# ----------------------
def read_file_safe(filename):
    try:
        with open(filename, "r") as f:
            return f.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")


print("\nReading python_notes.txt:")
print(read_file_safe("python_notes.txt"))

print("\nReading ghost_file.txt:")
print(read_file_safe("ghost_file.txt"))


# ----------------------
# Part C: Robust API Functions
# ----------------------

def fetch_products(limit=20):
    try:
        url = f"{BASE_URL}?limit={limit}"
        r = requests.get(url, timeout=5)
        r.raise_for_status()
        return r.json()["products"]

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)


def fetch_by_category(category):
    try:
        url = f"{BASE_URL}/category/{category}"
        r = requests.get(url, timeout=5)
        r.raise_for_status()
        return r.json()["products"]

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)


def create_product(payload):
    try:
        url = f"{BASE_URL}/add"
        r = requests.post(url, json=payload, timeout=5)
        r.raise_for_status()
        return r.json()

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)


# ----------------------
# Test API functions
# ----------------------
print("\nTesting API with Error Handling:\n")

products = fetch_products(5)
if products:
    for p in products:
        print(f"{p['title']} - ${p['price']}")


# ----------------------
# Part D: Input Validation Loop
# ----------------------
while True:
    user_input = input("\nEnter a product ID to look up (1–100), or 'quit' to exit: ")

    if user_input.lower() == "quit":
        print("Exiting program.")
        break

    # Validate number
    if not user_input.isdigit():
        print("Invalid input. Please enter a number.")
        continue

    product_id = int(user_input)

    # Validate range
    if product_id < 1 or product_id > 100:
        print("Please enter a number between 1 and 100.")
        continue

    try:
        url = f"{BASE_URL}/{product_id}"
        r = requests.get(url, timeout=5)

        if r.status_code == 404:
            print("Product not found.")
        else:
            r.raise_for_status()
            product = r.json()
            print(f"{product['title']} - ${product['price']}")

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)
```

```python
import requests
from datetime import datetime

BASE_URL = "https://dummyjson.com/products"
LOG_FILE = "error_log.txt"

# ----------------------
# Logger Function
# ----------------------
def log_error(function_name, error_type, message):
    with open(LOG_FILE, "a") as f:
        timestamp = datetime.now()
        f.write(f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n")


# ----------------------
# Trigger ConnectionError
# ----------------------
def test_connection_error():
    print("Triggering ConnectionError...\n")
    try:
        requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
    except requests.exceptions.ConnectionError as e:
        print("Connection failed. Please check your internet.")
        log_error("fetch_products", "ConnectionError", str(e))


# ----------------------
# Trigger HTTP 404 Error
# ----------------------
def test_http_error():
    print("\nTriggering HTTP 404...\n")
    try:
        r = requests.get(f"{BASE_URL}/999", timeout=5)

        if r.status_code != 200:
            print("Product not found.")
            log_error(
                "lookup_product",
                "HTTPError",
                "404 Not Found for product ID 999"
            )
        else:
            r.raise_for_status()

    except requests.exceptions.ConnectionError as e:
        print("Connection failed. Please check your internet.")
        log_error("lookup_product", "ConnectionError", str(e))

    except requests.exceptions.Timeout as e:
        print("Request timed out. Try again later.")
        log_error("lookup_product", "Timeout", str(e))

    except Exception as e:
        print("Error:", e)
        log_error("lookup_product", "Exception", str(e))


# ----------------------
# Display Log File
# ----------------------
def display_logs():
    print("\nError Log Contents:\n")
    try:
        with open(LOG_FILE, "r") as f:
            print(f.read())
    except FileNotFoundError:
        print("No log file found.")


# ----------------------
# Run Tests
# ----------------------
test_connection_error()
test_http_error()
display_logs()
```
